```python
import requests
```

```python
import json
```

```python
user = 'monicar@cnpq.br'
passw = 'cnpq.monicar'
```

# Pegando o token de acesso

```python
string = 'grant_type=&username='+user+'&password='+passw+'&scope=&client_id=&client_secret='
```

```python
headers = {'Content-Type': "application/x-www-form-urlencoded", 'Accept': "application/json"}
```

```python
api_url_login = "http://hom.api.programadegestao.economia.gov.br/auth/jwt/login"
```

```python
response = requests.post(api_url_login, headers=headers ,data=json.dumps(string))
```

```python
response
```

```python
rlogin_json = response.json()
```

```python
token = rlogin_json['access_token']
tipo =  rlogin_json['token_type']
```

```python
head = {'Authorization': 'Bearer {}'.format(token)}
```

```python
print(head)
```
